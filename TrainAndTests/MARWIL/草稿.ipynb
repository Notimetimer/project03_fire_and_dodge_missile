```python
# python
import numpy as np

# 一维向量
a = np.array([1,2,3])
b = np.array([4,5])

c = np.concatenate([a, b])   # array([1,2,3,4,5])
c2 = np.hstack((a, b))       # 同上
c3 = np.r_[a, b]             # 同上

# 列向量 (n,1)
a_col = a.reshape(-1,1)     # shape (3,1)
b_col = b.reshape(-1,1)     # shape (2,1)

c_col = np.vstack((a_col, b_col))            # shape (5,1)
# 或明确 axis
c_col2 = np.concatenate([a_col, b_col], axis=0)
```

```python
import os, sys
import pickle
def get_current_file_dir():
    # 判断是否在 Jupyter Notebook 环境
    try:
        shell = get_ipython().__class__.__name__  # ← 误报，不用管
        if shell == 'ZMQInteractiveShell':  # Jupyter Notebook 或 JupyterLab
            # 推荐用 os.getcwd()，指向启动 Jupyter 的目录
            return os.getcwd()
        else:  # 其他 shell
            return os.path.dirname(os.path.abspath(__file__))
    except NameError:
        # 普通 Python 脚本
        return os.path.dirname(os.path.abspath(__file__))

cur_dir = get_current_file_dir()

def load_il_and_transitions(folder, il_name, rl_name):
    if folder is None:
        # Notebook 中没有 __file__，使用当前工作目录
        folder = os.getcwd()
    il_path = os.path.join(folder, il_name)
    trans_path = os.path.join(folder, rl_name)
    il = None
    trans = None
    if os.path.isfile(il_path):
        with open(il_path, "rb") as f:
            il = pickle.load(f)
    if os.path.isfile(trans_path):
        with open(trans_path, "rb") as f:
            trans = pickle.load(f)
    return il, trans

# call from notebook


if __name__ == "__main__":
    il, trans = load_il_and_transitions(os.path.join(cur_dir, "IL"),\
                "il_transitions_combat.pkl",\
                    "transition_dict_combat.pkl")
    
    
    if il is None and trans is None:
        print("No il_transitions_chase.pkl or transition_dict_chase.pkl found in this directory.")
    else:
        if il is not None:
            print("Loaded il_transitions_chase.pkl -> keys:", list(il.keys()))
        if trans is not None:
            print("Loaded transition_dict_chase.pkl -> keys:", list(trans.keys()))
```
